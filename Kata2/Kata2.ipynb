## Objetivo Principal

Utilizar bibliotecas y planificar nuestro proyecto para crear programas.

### Aprenderemos a:

* Crear un proyecto de Python
* Desarrollar y ejecutar nuestro codigo de forma aislada en una maquina.
* Usar bibliotecas que otra persona haya escrito.
* Restaurar un proyecto a partir de una lista de dependencias.

#### Trabajar con paquetes 

La gran mayoria de los programas que escribamos se basarán en codigo escrito por alguien más. Este código a menudo viene en forma de paquetesm que son módulos externos o bibliotecas que se incluyen en el proyecto. Al igual que cualquier proyecto que requiera un conjunto de recursos, es importante considerar cómo nos aseguraremos de que los recursos adecuados estén disponibles para nuestro programa.

#### Entorno Virtual

Un entorno virtual es una copia autónoma de todo lo necesario para ejecutar el programa. Esto incluye el interprete de Python y cualquier biblioteca que necesitemos. Mediante el uso de un entorno virtual, podemos asegurarnos de que nuestro programa tendrá acceso a las versiones  recursos correctos para ejecutarse correctamente.

El flujo de trabajo se ve así:

* Crear un entorno virtual que no afecte al resto de la máquina.
* Ingresar al entorno virtual, donde especifique la versión de Python y las bibliotecas que necesita.
* Desarrollar nuestro programa.






#### Crear un entorno virtual

Para crear un entorno virtual, debemos llamar al módulo. El módulo espera un nombre como argunmento venv.

Desde nuestra terminal iremos al directorio donde queremos guardar el proyecto. (desktop/proyectos-azul/LaunchX/Kata2)

Usaremos el siguiente comando para llamar al modulo venv. 

***python3 -m venv env***

* python3 es la versión de python a utilizar.
* venv llavama al módulo venv (conocido como virtual environment)
* env es el nombre de nuestro entorno virtual

En este punto se crearán algunos directorios:

/env
    /bin
    /include
    lib

Nuestro entorno necesita el directoio venv para realizar el seguimiento de detalles como què version de Python y que bibliotecas está utilizando.
**NO colocaremos los archivos del programa en el directorio venv.** crearemos un directorio scr o similar.

La estructura hasta ahora se vé así:

/env
/src
    program.py

#### Activar el entorno virtual

En este punto ya tenemos el entorno virtual, pero no heos comenzado a usarlo. Para esto debemos actvarlo llamando **activate** a un script en el directorio env

source env/bin/activate

Ahora ya estamos dentro de nuestro entorno virtual, cualquier cosa que hagamos sucede de forma aislada.
